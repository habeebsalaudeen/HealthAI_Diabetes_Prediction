```python
import pandas as pd

df = pd.read_csv("pima_diabetes.csv", header=None)
df.head()
```

```python
df.columns = [
    "Pregnancies",
    "Glucose",
    "BloodPressure",
    "SkinThickness",
    "Insulin",
    "BMI",
    "DiabetesPedigreeFunction",
    "Age",
    "Outcome"
]

df.head()
```

```python
df.info()
```

```python
cols_with_zero = [
    "Glucose",
    "BloodPressure",
    "SkinThickness",
    "Insulin",
    "BMI"
]

for col in cols_with_zero:
    print(col, "zero values:", (df[col] == 0).sum())
```

```python
for col in cols_with_zero:
    df[col] = df[col].replace(0, df[col].median())
```

```python
for col in cols_with_zero:
    print(col, "zero values after cleaning:", (df[col] == 0).sum())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(9,7))
sns.heatmap(df.corr(), annot=True, fmt=".2f", cmap="viridis")
plt.title("Correlation Matrix")
plt.tight_layout()
plt.savefig("../04_Exports/visuals/correlation.png", dpi=300)
plt.show()
```

```python
df.hist(figsize=(12,10))
plt.tight_layout()
plt.savefig("../04_Exports/visuals/distributions.png", dpi=300)
plt.show()
```

```python
df.to_csv("pima_diabetes_cleaned.csv", index=False)
print("Cleaned dataset saved in 01_Data")
```
